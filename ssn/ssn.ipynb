```python
import pandas
names = pandas.read_csv('./ssn-names-1880-2015.csv')
```

```python
david_names = names[names['Name'] == 'David']
male_david_names = david_names[david_names['Gender'] == 'M']
%matplotlib inline
male_david_names.set_index('Year', inplace=True)
male_david_names['Count'].plot()
```

```python
female_david_names = david_names[david_names['Gender'] == 'F']
female_david_names.set_index('Year', inplace=True)
female_david_names['Count'].plot()
```

```python
robin_names = names[names['Name'] == 'Robin']
male_robin_names = robin_names[robin_names['Gender'] == 'M']
male_robin_names.set_index('Year',inplace=True)
male_robin_names['Count'].plot()
```

```python
names.groupby('Year')['Count'].sum().plot()
```

```python
cutter_names = names[(names['Name'] == 'Kathryn')]
cutter_names.set_index('Year',inplace=True)
cutter_names.plot()
```
